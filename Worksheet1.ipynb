1. Initialize an empty array with size 2X2.
2. Initialize an all one array with size 4X2.
3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5. Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}

```python
import numpy as np

empty_array=np.empty((2,2))
print("Empty array (2x2):"+'\n',empty_array)

ones_array=np.ones((4,2))
print("All ones array (4x2):"+'\n',ones_array)

fill_value = 6
new_array = np.full((2, 2), fill_value)
print("Fill array (2x2):"+'\n',new_array)

zeros_array = np.zeros_like(new_array)
print("Zeros array (2x2):"+'\n',zeros_array)

ones_like_array = np.ones_like(new_array)
print("Ones array (2x2):"+'\n',ones_like_array)

new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("Numpy array from list:", numpy_array)
```

```python
import numpy as np

a = np.array([[3,4],[7,8]])
b = np.array([[5,3],[2,1]])

a_inv = np.linalg.inv(a)

result = np.dot(a, a_inv)
rounded_result = np.round(result, decimals=10)

identity_matrix = np.eye(a.shape[0])
print("A * A^-1:\n", rounded_result )
print("Is A * A^-1 equal to the identity matrix?", np.allclose(rounded_result , identity_matrix))


# Perform matrix multiplication AB and BA
a_dot_b = np.dot(a,b)
print("\nA * B:\n", a_dot_b)

b_dot_a = np.dot(b,a)
print(" B* A:\n", b_dot_a)

print("Is AB equal to BA?", np.array_equal(a_dot_b, b_dot_a))


a_Transpose =np.transpose(a)
b_Transpose =np.transpose(b)
aT_bT = np.dot(a_Transpose,b_Transpose)
print("\n(A^T) * (B^T):\n", aT_bT)

ab_Transpose =np.transpose(a_dot_b)
print("(AB)^T:\n",ab_Transpose)

print("Is (AB)^T equal to B^T A^T?",np.array_equal(aT_bT,ab_Transpose))


A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B = np.array([-1, -3, 9])

# Solve for X using the inverse of A
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)

print("Solution X:\n", X)
```
